```python
import numpy as np
```

```python
x = 3*np.random.randn(100,1)
y = 3*4*x+np.random.randn(100,1)
```

```python
x_ = np.ones((100,1))
x = np.hstack((x_,x))
```

```python
def hypothesis(x,theta):
    return np.dot(x, theta)
```

```python
theta = np.ones((x.shape[1],1))
```

```python
def cost(x,y,m,theta):
    y_p = hypothesis(x,theta)
    cost = 1/m*np.sum(np.square(y_p,y))
    return cost
m = len(x)
cost(x,y,m,theta)
```

```python
def gd(x,y,m,theta,iters,lr):
    hist = []
    for i in range(iters):
        y_p = hypothesis(x,theta)
        theta -= 1/m*lr*np.sum(np.dot(x.T,(y_p-y)))
        hist.append(theta)
    return hist,theta
    
    
```

```python
hist,theta = gd(x,y,)
```

```python
def predict(x):
    return np.dot(x,theta)
```
