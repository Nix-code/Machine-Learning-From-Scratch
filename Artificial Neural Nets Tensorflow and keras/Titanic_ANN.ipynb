```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python

y = df["Survived"]

X = df.drop(['Survived'],axis=1)
X
```

```python
# now feature engineering
gender = pd.get_dummies("Sex")
```

```python
X.isna().sum()
```

```python
# lets make age as the range 

def age_feat(df):
    df.Age = df.Age.fillna(df.Age.mean())
    
    bins = [-1,0,5,15,25,35,50,70,100,120]
    cat = ["Invalid","infant","Children","Teens","Young","Adult","Elder","old","older"]
    
    age_me = pd.cut(df.Age, bins, labels=cat)
    df.Age = age_me
    return df

def last_Name(df):
    
    last = df.Name.apply(lambda x: x.split(' ')[1])
    df["SurName"] = last
    return df

def cabin_cat(df):
    df.Cabin = df.Cabin.fillna('N')
    cabin = df.Cabin.apply(lambda x: x[0])
    df.Cabin = cabin
    return df

def drop_feat(df):
    
    feat = ['PassengerId','Name','Ticket']
    df = df.drop(feat, axis=1)
    return df

def fare_transform(df):
    bins = [-1,0,7,14,31,512,1000]
    cat = ['Invalid','min','first_quartile','average','third_quartile','max']
    fare = pd.cut(df.Fare, bins, labels=cat)
    df.Fare = fare
    return df
def transform(df):
   
    df= age_feat(df)
    df = last_Name(df)
    df = drop_feat(df)
    df = cabin_cat(df)
    df = fare_transform(df)
    return df

X = transform(X)
```

```python
X.head()
```

```python
from sklearn.preprocessing import LabelEncoder

def encode_me(independent_variables):
    
    encode = LabelEncoder()

    feat = ['Sex','Age','Fare','Cabin','SurName','Embarked']

    for feature in feat:
        independent_variables[feature] = encode.fit_transform(independent_variables[feature])
    return independent_variables

    
```

```python
X = encode_me(X)
```

```python
import tensorflow as tf
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=23)
```

```python
## Make ANN model
```

```python
X_train.shape
X_train
```

```python
# import packages
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import ReLU
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
```

```python
model.add(Dense(units=9 , activation='relu'))
model.add(Dense(units=6, activation = 'relu'))
model.add(Dropout(0.2))
```

```python
model.add(Dense(units=8, activation = 'relu'))
model.add(Dropout(0.2))
model.add(Dense(units=1, activation = 'sigmoid'))
```

```python
model.compile(optimizer = 'adam', loss='binary_crossentropy',metrics=['accuracy'])
```

```python
hist = model.fit(X_train, y_train, validation_split=0.33, epochs=100, batch_size=10)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc = accuracy_score(y_pred, y_test)
```
