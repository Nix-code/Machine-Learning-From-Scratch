```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Elbow Method

```python
df = pd.read_csv('customer.csv')
X = df.iloc[:,3:].values
```

```python
fig = plt.figure(figsize = (10,8))
wcss = []

for i in range(1,11):
    model = KMeans(n_clusters = i, init='k-means++',max_iter = 300,n_init=10, random_state=0)
    model.fit(X)
    wcss.append(model.inertia_)
plt.plot(range(1,11),wcss)
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
model = KMeans(n_clusters = 5,init='k-means++', max_iter = 300, n_init=10, random_state = 0)
y_pred =model.fit_predict(X)
```

```python
y_pred
```

```python
fig = plt.figure(figsize = (12,8))

plt.scatter(X[y_pred == 0,0], X[y_pred ==0, 1], color = 'red', s=60, label = 'cluster1')
plt.scatter(X[y_pred == 1,0], X[y_pred ==1, 1], color = 'blue', s=60, label = 'cluster2')
plt.scatter(X[y_pred == 2,0], X[y_pred ==2, 1], color = 'green', s=60, label = 'cluster3')
plt.scatter(X[y_pred == 3,0], X[y_pred ==3, 1], color = 'yellow', s=60, label = 'cluster4')
plt.scatter(X[y_pred == 4,0], X[y_pred ==4, 1], color = 'purple', s=60, label = 'cluster5')

plt.scatter(model.cluster_centers_[:,0],model.cluster_centers_[:,1], color='magenta',s=100, label='Centroid')
plt.legend()
plt.xlabel("Spending score 1-100")
plt.ylabel("Annual Income (x$)")
plt.show()
```
